Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
# Glyphs are the geometrical shapes that get plotted on the charts
# These Shapes can be various markers, lines, bars, or even images
# Again same as the prevoius example
from bokeh.plotting import figure, show
import numpy as np
# Creat the plot with a title and axis labels
p = figure(title="Weight vs Helth", x_axis_label = "Weight", y_axis_label = "Helth")

# The data need to be plotted
weight = np.array([40,50,60,45,98,110])
heilth = np.array([0,0,1,0,2,3])

# Add a circule glyph for the scatter plot
p.circle(
    x = weight,
    y = heilth,
    size = 10, # Set marker size to 10 pixels
    color= "blue",
    alpha = 0.8 # Set marker transparency to 80%
 )

# Display the Plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Create a basic figure
p = figure(title="Customized Plot", x_axis_label="X Axis", y_axis_label="Y Axis")

# Customize axes
p.xaxis.axis_label = "Custom X Axis"
p.yaxis.axis_label = "Custom Y Axis"
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_color = "red"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"
p.xaxis.major_label_text_color = "orange"
p.yaxis.major_label_text_color = "green"
p.xaxis.major_label_text_font = "times"
p.yaxis.major_label_text_font = "helvetica"
p.xaxis.axis_line_width = 3
p.yaxis.axis_line_width = 2
p.xaxis.axis_line_color = "purple"
p.yaxis.axis_line_color = "brown"

# Customize title
p.title.text = "Customized Plot Title"
p.title.text_color = "darkblue"
p.title.text_font = "arial"
p.title.text_font_size = "20pt"
p.title.align = "center"

# Adding a renderer with legend
p.circle([1, 2, 3, 4], [4, 3, 2, 1], size=10, color="navy", alpha=0.5, legend_label="Data Points")

# Customize legend
p.legend.title = "Legend Title"
p.legend.title_text_font_size = "12pt"
p.legend.title_text_color = "blue"
p.legend.label_text_font_size = "10pt"
p.legend.label_text_color = "green"
p.legend.location = "top_left"
p.legend.orientation = "vertical"
p.legend.border_line_width = 2
p.legend.border_line_color = "black"
p.legend.background_fill_color = "lightgrey"

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

```python
"""
A Bokeh server is a component of the Bokeh visualization library that allows you
to build interactive web applications based on your Python visualizations.
"""

from bokeh.io import curdoc #Accesses the current document to add layouts and callbacks
from bokeh.models import ColumnDataSource # ColumnDataSource: A fundamental data structure in Bokeh for updating plots.
from bokeh.plotting import figure # Creates a new plot.
from bokeh.layouts import column #Arranges plots and widgets vertically.
from bokeh.models.widgets import Slider, Button #Widgets for user interaction.
import numpy as np

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=[], y=[])) # Initializes an empty data source.

# Create a figure
p = figure(title="Real-time Updating Plot", x_axis_label='x', y_axis_label='y')

# Add a line renderer
p.line('x', 'y', source=source)

# Define a callback to update data + Function to generate new data and update source
def update_data():
    new_data = dict(x=np.linspace(0, 4*np.pi, 200), y=np.sin(np.linspace(0, 4*np.pi, 200)) * np.random.random())
    source.data = new_data

# Add a button widget and configure its callback
button = Button(label="Update Data")
button.on_click(update_data)

# Add a slider widget and configure its callback
def update_plot(attr, old, new):
    scale = slider.value
    new_data = dict(x=source.data['x'], y=scale * np.sin(source.data['x']))
    source.data = new_data

slider = Slider(start=0.1, end=10, value=1, step=0.1, title="Scale")
slider.on_change('value', update_plot)

# Arrange plots and widgets vertically in layouts
layout = column(p, slider, button)

# Add the layout to the current document
curdoc().add_root(layout)

# Initial data update
update_data()


# Running the bokeh server in python file !!!
#  bokeh serve --show app.py
```
