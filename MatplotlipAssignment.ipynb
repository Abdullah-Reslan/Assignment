Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

scatter plot is used to plot each poin or intersection between two points.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x)) 

'''
. The np.random.normal() function generates random numbers from a normal distribution.
. The first argument (0) represents the mean of the normal distribution, which in this case is 0.
. The second argument (2) represents the standard deviation of the normal distribution, which in this case is 2.
. The third argument (50) specifies the number of random numbers to generate.
. Finally, 3 + ... adds 3 to each of the randomly generated numbers. So, each number in x will be shifted by 3.
'''
```

```python
plt.scatter(x,y, color="red", alpha=.9)
plt.title("This a graph using the normal distrubution")
plt.xlabel("this is the x axis")
plt.ylabel("this is the y axis")
plt.grid()
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function is used to display multiple plots within the same figure, allowing for better organization and comparison of data

```python
# Use the following data:

# For line 1
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,10,20,300,400,500])

# For line 2
x2= np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

# For line 3
x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

# for line 4
x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])
```

```python
# First subplot
plt.subplot(2,2,1)
plt.plot(x1 ,y1 , "r-")

# second subplot
plt.subplot(2,2,2)
plt.plot(x2,y2, "b-")

# third subplot
plt.subplot(2,2,3)
plt.plot(x3,y3, "g-")

# fourth plot
plt.subplot(2,2,4)
plt.plot(x4, y4, "m-")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for each line plot
data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))  # Set the figure size

for i, (x, y) in enumerate(data, start=1):
    plt.subplot(2, 2, i)  # Select the subplot
    plt.plot(x, y, label=f'Line {i}')  # Plot the data
    plt.xlabel('X')  # Set X-axis label
    plt.ylabel('Y')  # Set Y-axis label
    plt.title(f'Line {i} Plot')  # Set title for the subplot
    plt.legend()  # Show legend

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()  # Display the figure with all subplots
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a graphical representation of categorical data where the length of bars <br>represents the frequency, count, or any other measure associated with each category.<br> It's commonly used to compare the values of different categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(5,3))
plt.bar(company, profit)
plt.xlabel("representing the Company name")
plt.ylabel("representing the profit")
plt.title("Profit and companys")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.barh(company, profit)
plt.ylabel("representing the Company name")
plt.xlabel("representing the profit")
plt.title("Profit and companys")
plt.grid()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers in a compact and easy-to-understand format.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.xlabel("Dataset")
plt.ylabel("values")
plt.title("Box plot with two Datasets")
plt.xticks([1,2], ["Dataset1", "Dataset2"]) # Adding custom labels for the datasets
plt.show()
```
